```python
clean_listings = []
for i in listings:
    clean_listings.append([(str(i)),(listings[i].isnull().sum() / (len(listings[i])))*100])

clean_listings
df = pd.DataFrame(clean_listings, columns=['Column','Percentage Null Values'])
null_columns = df.loc[df['Percentage Null Values'] != 0]
null_columns
```

```python
def create_buckets(name, df):
    range_list ={'0-500':0,'501-1000':0,'1001-1500':0,'1501-2000':0,'above 2000':0}
    for i in list(df[name].unique()):
        temp = i.split('$')
        num = float(temp[1].replace(',',''))
#         print num
        if num >= 0.0 and num <= 500.0:
            range_list['0-500'] += 1
        elif num >=501.0 and num <= 1000.0:
            range_list['501-1000']+=1
        elif num >=1001.0 and num <= 1500.0:
            range_list['1001-1500']+=1
        elif num >=1501.0 and num <= 2000.0:
            range_list['1501-2000']+=1
        else:
            range_list['above 2000']+=1
    return range_list

def create_pie(dict1):
    labels = dict1.keys()
    fig1, ax1 = plt.subplots()
    ax1.pie(dict1.values(),labels=labels,shadow=True, startangle=90)
    plt.show()
    

```

```python

create_pie(create_buckets('price',listings))
```
